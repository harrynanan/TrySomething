```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar, Pie, Page, Funnel
from pyecharts.globals import ThemeType
```

```python
df1 = pd.read_csv(r'F:\data_files\Task_4_lianjia_sale.csv', encoding='utf-8')
df1.head(5)
```

```python
data_price = df1.groupby('区域')['价格'].agg([('区域平均价格','mean')])
df_avg = round(data_price.sort_values(by='区域平均价格'))
df_avg.head()
```

```python
x1 = df_avg.index.values.tolist()
y1 = df_avg['区域平均价格'].tolist()
bar = Bar()
bar.add_xaxis(x1)
bar.add_yaxis("每平米价格/元",y1)
bar.set_global_opts(title_opts=opts.TitleOpts(title="上海各区二手房平均价格"))
bar.reversal_axis()
bar.render_notebook()
```

```python
df_decoration = df1.groupby('装修类型')['装修类型'].agg([('房源数量','count')])
df_decoration
```

```python
x2 = df_decoration.index.values.tolist()
y2 = df_decoration['房源数量'].values.tolist()
c = (
Pie()
.add("", [list(z) for z in zip(x2,y2)],radius=["40%", "75%"])
.set_global_opts(title_opts=opts.TitleOpts(title="装修类型比例"), 
                 legend_opts=opts.LegendOpts(
                orient="vertical",#设置标签垂直
                pos_top="15%", pos_left="2%"#设置标签位置
            ))
.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))#格式化标签输出内容
)
c.render_notebook()
```

```python
import pandas as pd
import numpy as np
from pyecharts.charts import Map,Geo
from pyecharts import options as opts
from pyecharts.globals import ThemeType
import tushare as ts
df1 = pd.read_csv(r'F:\data_files\Task_4_lianjia_sale.csv', encoding='utf-8')
df_avg = round(df1.groupby('区域')['价格'].agg([('区域平均价格','mean')]).sort_values(by='区域平均价格'))
df_avg.head()
```

```python
district_name = [i+'区' if i!='浦东' else i+'新区' for i in df_avg.index.tolist()]
district_price = df_avg['区域平均价格'].tolist()
district_data = [[district_name[i],district_price[i]] for i in range(len(district_name))]
map_shanghai = Map()
map_shanghai.set_global_opts(
    title_opts=opts.TitleOpts(title="上海各区二手房平均房价"),
    visualmap_opts=opts.VisualMapOpts(max_=140000)
)
map_shanghai.add("每平米/元",district_data,maptype='上海')
map_shanghai.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
map_shanghai.render_notebook()
```

```python
df2 = ts.get_area_classified()
company = df2.groupby('area')['area'].agg([('上市公司数量','count')])
province_name = company.index.values.tolist()
province_ccount = company['上市公司数量'].values.tolist()
province_data = [[province_name[i],province_ccount[i]] for i in range(len(province_name))]
map_china=Map()
map_china.set_global_opts(
    title_opts=opts.TitleOpts(title="中国各省上市公司数量"),
    visualmap_opts=opts.VisualMapOpts(max_=500)
)
map_china.add("上市公司数量",province_data,maptype='china')
#map_china.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
map_china.render_notebook()
```
